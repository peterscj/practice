```python
def main():
    print("Welcome to cole's shitty tic tac toe game!")
    update_board()
    print("end")

def display_board(grid):
    
    # Update variables to be type dict for easier tracking of grid/value pairs -- DONE
    
    t = list(grid.keys())
    
    bars_bottom = "||_____||_____||_____||"
    bars_top = "||     ||     ||     ||"
    
    # Available grid
    bars_mid_3 = "||  {0}  ||  {1}  ||  {2}  ||".format(grid[7],grid[8],grid[9]) 
    bars_mid_2 = "||  {0}  ||  {1}  ||  {2}  ||".format(grid[4],grid[5],grid[6])
    bars_mid_1 = "||  {0}  ||  {1}  ||  {2}  ||".format(grid[1],grid[2],grid[3])

    # Show grid mapping
    _bars_mid_3 = "||  {0}  ||  {1}  ||  {2}  ||".format(t[6],t[7],t[8]) 
    _bars_mid_2 = "||  {0}  ||  {1}  ||  {2}  ||".format(t[3],t[4],t[5])
    _bars_mid_1 = "||  {0}  ||  {1}  ||  {2}  ||".format(t[0],t[1],t[2])

    
    print("        Mapping","     ","             Current Board\n\n")
    print(bars_top,"     ",bars_top)
    print(_bars_mid_3,"     ",bars_mid_3)
    print(bars_bottom,"     ",bars_bottom)
    print(bars_top,"     ",bars_top)
    print(_bars_mid_2,"     ",bars_mid_2)
    print(bars_bottom,"     ",bars_bottom)
    print(bars_top,"     ",bars_top)
    print(_bars_mid_1,"     ",bars_mid_1)
    print(bars_top,"     ",bars_top)

    
def update_board():
    
    x = 0
    player = 'O'
    new_grid =  {i:" " for i in range(1,10)} # intialize empty grid
    
    display_board(new_grid)
    
    end_game = False
    while not end_game and x < 9:
        
        # X always goes first and set player equal to current turn
        turn = next_turn(player) 
        player = turn 
        
        # Get the grid ID from the player and then update the grid 
        # with the player's symbol
              
        
        
        ## Verify that the player's input is valid
        
        attempt = False
        while not attempt:

            # Controls to ensure that an integer is entered
            # If an exception is raised, loop will break and restart
            try:
                move = int(input("Player {0} may choose a box ID: ".format(turn)))
            
            except ValueError:
                print('Could not convert to an integer!')
                break
                
            except:
                print("Error. Plz input an integer between 1-9")
                break
                
            if move > 9:
                print("You must input a value less than or equal to 9!!\n")
                move = int(input("Player {0} may choose a box ID: ".format(turn)))

            elif move in [i for i in new_grid if new_grid[i] == ' ']:
                new_grid[move] = player # Update grid with X or O
                attempt = True    
            else:
                print("You must select an empty grid")
                move = int(input("Player {0} may choose a box ID: ".format(turn)))
        
        
        # Check for available moves and have player re-enter move if they selected a full box
            
            x += 1 # Adjust x

            display_board(new_grid) # Display the board

            # Check if one of the two players has won
            x_wins = game_over(new_grid,'X')
            o_wins = game_over(new_grid,'O')

            if x_wins or o_wins == True:
                end_game = True

    if end_game:
        print("The winner is", turn)
    else:
        print("The game ended in a tie...like it always should")

def next_turn(prev):
    if prev == "O" or prev == "":
        return "X"
    
    else:
        return "O"
    
def game_over(current_grid,player):

    """
    Accepts markup of the current grid and a player to check if that player has won any
    of the winning combinations
    
    """
    winning_combos = [[1,2,3,'x'],[4,5,6,'x'],[7,8,9,'x'], # Horizontal combinations
                      [1,4,7,'x'],[2,5,8,'x'],[3,6,9,'x'], # Vertical combinations
                      [1,5,9,'x'],[3,5,7,'x']] # Diagonal combinations
    
    moves = [i for i in current_grid if current_grid[i] == player] # Confirmed works

    for combo in winning_combos: # Loop through all combos and eliminate any winning moves that the player has made
        for move in moves:
            if move in combo:
                combo.pop(combo.index(move))
    
    # Check if any of the combinations have been reduced to length 1
    # Returns boolean value depending on if game should continue to next
    # turn
    
    for combo in winning_combos:
        if len(combo) > 1:
            continue
        else:
            return True # Winning condition

    return False # play on condition

if __name__ == "__main__":
    main()
    
```
